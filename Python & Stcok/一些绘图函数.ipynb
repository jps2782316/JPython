```python
import tushare as ts
import pandas as pd

#1、使用tushare获取某只股票的历史行情数据

df = ts.get_k_data('600519', start='2015-01-01')
df.to_csv('maotai_600519.csv') #存储数据
df

```

```python
#读取数据  把date变成索引列，并且把它变成时间对象
df = pd.read_csv('maotai_600519.csv', index_col='date', parse_dates=['date'])
#花式索引，只保留必要数据
df = df[['open', 'close', 'high', 'low']] 
df
```

```python
#2、输出该股票所有收盘比开盘上涨3%以上的日期

#布尔索引过滤
df[(df['close'] - df['open']) / df['open']>=0.03].index
```

```python
#3、输出该股票所有开盘比前日收盘跌幅超过2%的日期

#除索引列外，其他数据列往下移一行
#df.shift(1) 
bools = (df['open']-df['close'].shift(1)) / df['close'].shift(1) <= -0.02
df[bools]
```

```python
#4、假如我从2010年1月1日开始，每月第一个交易日买入一手股票，每年最后一个交易日卖出所有股票，到今天为止，我的收益如何？

#剔除首尾无用数据，比如2001年8月的数据是之后后半个月，没有前半个月的，不符合第一个交易日买入的需求，所以剔除。
#df = df['2001-09':'2017-11']
#按月重新取样，算出这个月的平均值
#df.resample('M').mean()
df_monthly = df.resample('M').first()
df_yearly = df.resample('A').last()[:-1] #切掉最后一行，因为2021年还没到最后最后一个月，不满足每年最后一个交易日卖出的条件
df_yearly

#取出2015的所有数据
#df_monthly['2015']

#股票当前价格
price_last = df['open'][-1]
#买入股票总成本 花费
total_cost = 0 
#目前持股数
total_hold = 0 
for year in range(2015, 2022):
    print('年份%d' %year)
    month_opens = df_monthly[str(year)]['open']
    #当前这一年花费：当前年所有月的开盘价之和，然后乘以一手(100股)
    current_cost = month_opens.sum() * 100
    total_cost += current_cost
    #当前这一年买了多少股: 购买次数 * 1手
    current_hold = len(month_opens) * 100
    total_hold += current_hold
    if year != 2021:
        #当前年最后一个交易日卖出所得金额
        get_money = df_yearly[str(year)]['open'][0] * total_hold
        total_cost -= get_money
        #清空所有持股
        total_hold = 0
        print('年份', year, '累计成本', total_cost)
    
#目前持股价值
hold_money = total_hold*price_last
print('未清仓持股价值', hold_money, '已清仓成本', total_cost)
earnings = -total_cost + hold_money
print('赚了%.1f' %earnings)

```
